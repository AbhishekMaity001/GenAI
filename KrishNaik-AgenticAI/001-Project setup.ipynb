##### Install the UV package manger
         Step 1 : install uv using below command
                  powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
         Step 2:  Initialize a working repository
                     You might get some error, so set the environment variable first, before running the uv init command
                         $env:Path = "C:\Users\Abhishek Maity\.local\bin;$env:Path"
                     
                     ## for installing python version 
                         - uv python install 3.13

                     ## Pin it for your current project:
                          - uv python pin 3.13
                          
                     ## initialize uv
                     uv init
                    - some folders will automatically get created like pyproject.toml file, readme.md file etc.
                    
         Step 3: Create a virtual environment
                 uv venv --python 3.13
                 
         Step 4: Activate the environment with the bellow command
                 .venv\Scripts\activate
                 
         Step 5: create a requirements.txt file and include the requirements

         Step 6: uv add -r requirements.txt
