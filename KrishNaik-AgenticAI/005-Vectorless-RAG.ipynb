# 📑 What is PageIndex?

        PageIndex is a vectorless, reasoning-based RAG (retrieval) framework that mirrors how humans read long, complex documents. Instead of vector similarity search, it transforms documents into a tree structure, and enables LLMs to perform agentic reasoning over that tree for context-aware retrieval. The retrieval is traceable and explainable, with no vector databases and no chunking.
