### One Hot Encoding

```python
corpus = ["i love nlp", "i learn gen ai", "i am working with accenture"]
corpus
```

```python
unique_words = list(set(" ".join(corpus).split()))
unique_words
```

```python
word_to_index = {v:k for k,v in dict(enumerate(unique_words)).items()}
word_to_index
```

```python
one_hot_encoder = []
for sentence in corpus:
    # print(sentence.split())
    sentence_vector = []
    for word in sentence.split():
        vector = [0] * len(unique_words)
        vector[word_to_index[word]] = 1
        # print(vector)
        sentence_vector.append(vector)
    # print(sentence_vector)
    
    one_hot_encoder.append(sentence_vector)
print(one_hot_encoder)
```

### Text Pre processing function

```python
import re
def word_preprocessing(text):
    text = text.lower()  # lower case
    text = re.sub(r'\d+', "", text) # remove special character
    text = text.translate(str.maketrans("", "", string.punctuation))
    word = word_tokenize(text)
    word = [x for x in word if x not in stopwords.words("english")]
    return word
       
```

```python
corpus
```

```python
processed_corpus = [word_preprocessing(text) for text in corpus]
# [text for text in corpus]
processed_corpus
```

```python
from gensim.models import Word2Vec
```

```python
word2vec_cbow = Word2Vec(sentences=processed_corpus, vector_size=100, window=5, min_count=1, sg=0)
word2vec_skipgram = Word2Vec(sentences=processed_corpus, vector_size=100, window=5, min_count=1, sg=1)

# window size = distance between target word & context word (window size is larger then more symmentic relation but window size is smaller then  more syntactic relationship)
# min_count = consider words this no. of time
# sg parameter = 0 means CBOW, cbow is used to generate the target word, to generate surrounding words use sg=1, means skip gram
```

```python
word2vec_cbow, word2vec_skipgram
```

```python
word2vec_cbow.wv["maity"]
```

```python
word2vec_cbow.wv.most_similar("nlp", topn=5), word2vec_skipgram.wv.most_similar("nlp" , topn=5)
```

```python
word2vec_cbow.wv.similarity("nlp", "maity")
```
