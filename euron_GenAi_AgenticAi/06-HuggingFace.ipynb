## Importing the pre-trained model from Transformers

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"


tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True, use_safetensors=False)
```

```python
# text = "i love all the features provided by euron"
text = "My name is Abhishek maity and i am a data scientist"
```

```python
input = tokenizer(text, return_tensors="tf")
input
```

```python
output = model(**input)
output
```

```python
prob = tf.nn.softmax(output.logits, axis=-1).numpy()[0]
prob
```

```python
predict_class_id = np.argmax(prob)
predict_class_id
```

```python
model.config.id2label[predict_class_id]
```

## Loading IMDB Dataset for finetuning LLM Model

```python
from datasets import load_dataset
dataset = load_dataset("imdb")
dataset
```

### Tokenizing the full imdb dataset

```python
from transformers import AutoTokenizer

model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

def tokenizer_fn(data):
    return tokenizer(data["text"], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenizer_fn, batched=True)
```

```python
tokenized_dataset
```

#### We don't need text column, as we have already tokenized all the texts

```python
tokenized_dataset = tokenized_dataset.rename_column("label", "labels")
tokenized_dataset = tokenized_dataset.remove_columns("text")
tokenized_dataset
```

##### setting the format to Pytorch format

```python
tokenized_dataset.set_format("torch")
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=2)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./result", 
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=1,
    weight_decay=0.01
)
```

```python
!pip install transformers[torch] accelerate>=0.26.0
```

### Starting the Fine Tuning of the LLM

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"].select(range(100)),
    eval_dataset=tokenized_dataset["test"].select(range(100))
)

trainer.train()
```
